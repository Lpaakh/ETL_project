# ETL Project - COVID-19 cases and PPP loans in California

```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# US county file

us_county_file = "us_county.csv"
us_county_df = pd.read_csv(us_county_file)
us_county_df.head()
```

```python
us_county_df_copy = us_county_df.copy()
us_county_df_copy.head()
```

```python
# drop all states that are not california
ca_county_df = us_county_df_copy.loc[us_county_df["state_code"] == "CA", :]
ca_county_df.head()
```

```python
clean_county = ca_county_df.county.map(lambda x: x.replace(" County", ""))
ca_county_df.county = clean_county
ca_county_df.head()
```

```python
# export a clean csv
ca_county_df.to_csv('ca_county_clean.csv')
```

### Alison's part

```python
# US county covid file

covid_us_county_file = "covid_us_county.csv"
covid_us_county_df = pd.read_csv(covid_us_county_file)
covid_us_county_df.head(15)
```

```python
# Cleaning and making a copy

covid_us_county_cols = ["fips", "county", "state", "date", "cases", "state_code", "deaths"]
covid_us_county_transformed= covid_us_county_df[covid_us_county_cols].copy()
covid_us_county_transformed.head() 
```

```python
# Include only california and sum the 
ca_covid_county = covid_us_county_transformed.loc[covid_us_county_df["state_code"] == "CA", :]
ca_covid_county 
ca_covid_county_grouped = ca_covid_county.groupby(["fips","county", "state","state_code"]).sum()
ca_covid_county_grouped.head()
ca_covid_county_clean = ca_covid_county_grouped.reset_index()
ca_covid_county_clean.head()
```

```python
# export a clean csv
ca_covid_county_clean.to_csv('ca_covid_county_grouped.csv')
```

### Laura's part

```python
# zip, county and fips file

zip_county_fips_file = "zip_county_fips.csv"
zip_county_fips_df = pd.read_csv(zip_county_fips_file)
zip_county_fips_df.head()
```

```python
# Split the county name
clean_county = zip_county_fips_df.COUNTYNAME.map(lambda x: x.replace(" County", ""))
zip_county_fips_df.COUNTYNAME = clean_county
zip_county_fips_df
```

```python
# drop all states that are not california
cleaned_df = zip_county_fips_df.loc[zip_county_fips_df["STATE"] == "CA", :]
cleaned_df.head()

cleaned_zip_county_fips_df = cleaned_df.rename(columns = {'STCOUNTYFP': 'fips', "STATE": "state", "COUNTYNAME": "county", "CITY": "city", "ZIP":"zip", "CLASSFP":"classfp"})
cleaned_zip_county_fips_df
```

```python
# export as csv
cleaned_zip_county_fips_df.to_csv("cleaned_zip_county_fips.csv")
```
